```python
timelapse = geemap.landsat_timelapse(
    fire.geometry().bounds().buffer(8000),
    out_gif='landsat.gif',
    start_year= 2016,
    end_year= 2016,
    start_date= '09-01',
    end_date= '12-01',
    frequency='month',
    bands=['SWIR2','NIR', 'Red'],#nir, swir2, gr
    frames_per_second=1,
    title='Landsat-8 Timelapse',
)
geemap.show_image(timelapse)
```
